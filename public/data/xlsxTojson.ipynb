```python
import pandas as pd
```

```python
file = './busData.xlsx'
```

```python
with open(file, 'rb') as f:
    df = pd.read_excel(f)
    df.rename(columns={"정류소명":"STATION_NAME", "X좌표": "Lng", "Y좌표": "Lat"}, inplace=True)
    df.to_json('./busData.json', orient='records', force_ascii=False, index=4)
```
